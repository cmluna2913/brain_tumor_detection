```python
# for data path
import configparser

# basics
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

# image importing
import os
import cv2
```

```python
# importing config.ini to get data path
config = configparser.ConfigParser()
config.read("src/config.ini") 

path = config['paths']['data']
print(path)
```

```python
# folder of no brain tumor images
no_bt = path+"no/"
# folder of yes brain tumor images
yes_bt = path+"yes/"

#def import_images(path):
# iterate through images in folder
for img in os.listdir(no_bt):
    # create image path
    img_path = no_bt+img
    # import and make grayscale
    img_arr = cv2.imread(img_path)
    img_arr = cv2.cvtColor(img_arr, cv2.COLOR_BGR2GRAY)

    # get edges of brain in images
    edge = cv2.Canny(img_arr, 50,150)
    contours, hierarchy = cv2.findContours(edge, 
    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    plt.imshow(edge)
    plt.show()
    
    # get dimension for cropping image to just the brain
    x,y,w,h = cv2.boundingRect(edge)
    plt.imshow(cv2.rectangle(img_arr, (x-1,y-1), (x+w+1, y+h+1), (255,0,0),3))
    plt.show()

    # crop image
    cropped = img_arr[y:y+h, x:x+w]
    plt.imshow(cropped)
    plt.show()

    # only want to test on first one
    break
```
